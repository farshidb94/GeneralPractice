```python
import numpy as np
import matplotlib.pyplot as plt


x = np.linspace(0,10,100)

for i in np.linspace(0,1,50):
  plt.plot (x, x**i, color=[0.3,0.5,0.14])

plt.show()
```

```python
fig, ax = plt.subplots()
y = x**2
ax.plot(x,y, 'm-')
ax.annotate('here', (x[20],y[30]), arrowprops= dict(), xycoords='data', xytext=(5,5))
plt.show()
```

```python

y = x**2 + np.cos(x)*10

#minpnt = np.argmin(y)
fig, ax = plt.subplots()
ax.plot(x,y, 'ko-', markersize = 15, markerfacecolor = 'teal')
ax.annotate(
    'min:', (x[4], y[0]),
            arrowprops={'arrowstyle':'wedge', 'color': 'm', 'lw' : 1.5},
            xycoords='data',
            xytext=(5,5) )
ax.set_ylabel('exercise')
plt.show()
```
